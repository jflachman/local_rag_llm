# Sentence Transformers Embedding model

- Sentence Transformers: https://sbert.net/
- Quick Start: https://www.sbert.net/docs/quickstart.html#sentence-transformer
- Model files: https://huggingface.co/sentence-transformers/all-mpnet-base-v2

```python
from sentence_transformers import SentenceTransformer
modelPath = "/mnt/c/ML/DU/local_rag_llm/models/sentence-transformers/all-MiniLM-L6-v2"

# ---- Uncomment the following two lines to download the model the first time.
# model = SentenceTransformer('all-MiniLM-L6-v2')
# model.save(modelPath)
model = SentenceTransformer(modelPath)
# # 1. Load a pretrained Sentence Transformer model
# model = SentenceTransformer(modelPath)

# The sentences to encode
sentences = [
    "The weather is lovely today.",
    "It's so sunny outside!",
    "He drove to the stadium.",
]

# 2. Calculate embeddings by calling model.encode()
embeddings = model.encode(sentences)
print(embeddings.shape)
# [3, 384]

# 3. Calculate the embedding similarities
# similarities = model.similarity(embeddings, embeddings)
# print(similarities)
# tensor([[1.0000, 0.6660, 0.1046],
#         [0.6660, 1.0000, 0.1411],
#         [0.1046, 0.1411, 1.0000]])
```
